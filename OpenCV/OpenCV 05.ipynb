```python
import cv2
```

```python
cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter_fourcc(*'XVID')        #fourcc = cv2.VideoWriter_fourcc('X','V','I','D')
out = cv2.VideoWriter('Saved_Video.avi', fourcc, 10.0, (640,480))

if not cap.isOpened():
    print("Device Not Found.")
while cap.isOpened():
    ret, frame = cap.read()
    
    if ret:
        out.write(frame)
        cv2.imshow('frame', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
        
cap.release()
out.release()
cv2.destroyAllWindows()
```
