```python
File_Name = str(input("Please Enter the Name of your File: "))
print("-------------------------------------------------")
u_words = []
T_word_count = 0
oc_words = {}

Filter_list = ["!", "@", "#", "$", "%",
               "^", "&", "*", "(", ")",
               ",", "`", "~", "?", ">",
               "<", ":", ";", "'", "=",
               "-", ".", "\n"]

###### Start of Functions!
def SpaceRemover(line):
    pre = ""
    NewLine = ""
    for char in line:
        if char == " ":
            if pre != " ":
                pre = " "
                NewLine += char
            else:
                pre = " "
        else:
            NewLine += char
            pre = char
    return NewLine

def Filtering(line):
    NewLine = ""
    for char in line:
        if char not in Filter_list:
            NewLine += char
    return NewLine

def UniqueWordValidator(line):
    line = line.split(" ")
    for w in line:
        if w not in u_words:
            u_words.append(w)

def LastSpaceRemover(line):
    if line.endswith(" "):
        lineSize = len(line)
        NewLine = line[:(lineSize-1)]
        return NewLine
    else:
        return line
def WordCounter(line):
    my_line = line.split(" ")
    return len(my_line)
def WordInTextcounter(line):
    words = line.split(" ")
    for word in words:
        if word in oc_words:
            oc_words[word] +=1
        else:
            oc_words[word] = 1
def WordPercentage(dict_ofWords):
    newDict = {}
    fileLenght = LenghtofFile_inWords(oc_words)
    for key in dict_ofWords:
        newDict[key]=(oc_words[key]/fileLenght)
    return newDict
    
def LenghtofFile_inWords(dict_ofWords):
    sum = 0
    for key in dict_ofWords:
        sum += oc_words[key]
    return sum

###### End of Functions!
try:
    with open(File_Name+".txt", "r") as myFile:
        for line in myFile:
            line = SpaceRemover(line)
            line = Filtering(line)
            line = LastSpaceRemover(line)
            T_word_count += WordCounter(line)
            WordInTextcounter(line)
            UniqueWordValidator(line)
except FileNotFoundError:
    print("File Not Found!")
except:
    print("An Error occured!")

print("\nLenght of File ( Number of total words in file): "+str(LenghtofFile_inWords(oc_words))+"\n")
print("-------------------------------------------------")
print("\nList of All Worlds: \n{0}".format(u_words))
print("-------------------------------------------------")
print("\nList of All Worlds with their Number of repeatition: \n{0}\n".format(oc_words))
print("-------------------------------------------------")
oc_words = WordPercentage(oc_words)
print("\nList of All Worlds with their occupation percentage: \n")
sum=0
for key in oc_words:
    print("{x} => {y}".format(x=key,y=oc_words[key]))
    sum+=oc_words[key]
print("-------------------------------------------------")
print("Sum: {x}".format(x=sum))
```
