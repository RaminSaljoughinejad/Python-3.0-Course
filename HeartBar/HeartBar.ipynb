```python
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.image as mpimg
```

```python
plt.rcParams['figure.figsize'] = (9, 1)
fileName = input("Jpg file: ") + ".jpg"
im = Image.open(fileName)
img = mpimg.imread(fileName)
#split = int(input("Max length limit: "))
split = 0
pix = im.load()
xP = []
yP = []
ly = 0
lx = 0
for x in range(0, im.size[0]):
    p = pix[x, 0]
    py = ly
    for y in range(0, im.size[1]):
        t = pix[x, y]
        if t[0] + t[1] + t[2] < p[0] + p[1] + p[2] and 800 > p[0] + p[1] + p[2]:
            p = t
            py = y
    xP.append(x)
    yP.append(im.size[1] - py)
    ly = py
    if split > 0:
        if (not x % split and x > 0) or x == im.size[0] - 1:
            plt.plot(xP, yP)
            plt.title(str(lx) + " - " + str(x))
            plt.show()
            xP = []
            yP = []
            lx = x
if split <= 0:
    plt.plot(xP, yP)
    plt.title("0 - " + str(len(xP)))
    plt.ylim(0, im.size[1])
    plt.xlim(0, im.size[0])
    plt.show()
imgplot = plt.imshow(img)
```
