```python
import pygame
import random
```

```python
def random_gap(edge,mini_gap,maxi_gap):
    return random.randint(edge+mini_gap,edge+maxi_gap)

def frame(game_x,jump):
    if 0<=game_x%10<5:
        if jump:
            gameDisplay.blit(trex_jump,(trex_x,trex_y))
        else:    
            gameDisplay.blit(trex1,(trex_x,trex_y))
    elif 5<=game_x%10<10:
        if jump:
            gameDisplay.blit(trex_jump,(trex_x,trex_y))
        else:
            gameDisplay.blit(trex2,(trex_x,trex_y))
    if game_x == 1000:
        game_x = 1
    else:
        game_x += 1
    return game_x
```

```python
pygame.init()

main_widnow_width = 1000
main_widnow_height = 500

min_gap = 0
max_gap = 50

gameDisplay = pygame.display.set_mode((main_widnow_width,main_widnow_height))
pygame.display.set_caption("Trex")

#Color
white = (255,255,255)

#Trees
tree_image = [pygame.image.load("tree_mix2.png"),
              pygame.image.load("tree1s1.png"),
              pygame.image.load("tree1s2.png"),
              pygame.image.load("tree1s3.png"),
              pygame.image.load("tree2s1.png"),
              pygame.image.load("tree2s2.png"),
              pygame.image.load("tree2s3.png"),
              pygame.image.load("tree3s1.png"),
              pygame.image.load("tree3s2.png"),
              pygame.image.load("tree3s3.png"),
              pygame.image.load("tree4s1.png"),
              pygame.image.load("tree4s2.png"),
              pygame.image.load("tree4s3.png"),]
 
trees_y = [295,300,320,345,300,320,345,300,320,345,300,320,345]

trees_width = [145,86,73,60,86,73,60,86,73,60,86,73,60]
trees_height = [130,130,110,90,130,110,90,130,110,90,130,110,90]

trees_location = [1000,1400,1800,2200]

trees_cur = [random.randint(0,len(tree_image)-1),
             random.randint(0,len(tree_image)-1),
             random.randint(0,len(tree_image)-1),
             random.randint(0,len(tree_image)-1)]
trees_cur_y = [trees_y[trees_cur[0]],
               trees_y[trees_cur[1]],
               trees_y[trees_cur[2]],
               trees_y[trees_cur[3]]
              ]
print(trees_cur)
print(trees_cur_y)

trees_x = [random.randint(trees_location[0]+min_gap,trees_location[0]+max_gap),
           random.randint(trees_location[1]+min_gap,trees_location[1]+max_gap),
           random.randint(trees_location[2]+min_gap,trees_location[2]+max_gap),
           random.randint(trees_location[3]+min_gap,trees_location[3]+max_gap)
          ]




#Trex
trex1 = pygame.image.load("trex_run1.png")
trex2 = pygame.image.load("trex_run2.png")
trex_jump = pygame.image.load("trex_jump.png")
trex_x = 50
trex_y = 300
trex_width = 233
trex_height = 216

#Ground
ground1 = pygame.image.load("desert.png")
ground_x1 = 0
ground_y1 = 350
ground2 = ground1
ground_x2 = 1000
ground_y2 = ground_y1

clock = pygame.time.Clock()

game_x = 1


jump = False
jump_i = -30

gamestop = False
gameSpeed = 5
alive = True
while alive:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            alive = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE and gamestop != True:
                if jump == False:
                    jump = True
    if jump:
        if jump_i<=30:
            trex_y = (main_widnow_height-trex_height)- (180 - ((jump_i**2) / 5))
            jump_i+=1 
        else:
            jump = False
            jump_i = -30
    
    gameDisplay.fill(white)
    gameDisplay.blit(ground1,(ground_x1,ground_y1))
    gameDisplay.blit(ground2,(ground_x2,ground_y2))
    
    for i in range(0,4):
        gameDisplay.blit(tree_image[trees_cur[i]],(trees_x[i],trees_cur_y[i]))
        if trees_x[i]<-150:
            last_tree = max(trees_x)
            trees_x[i] = random.randint(1400+min_gap,1400+max_gap)
            trees_cur[i] = random.randint(0,len(tree_image)-1)
            trees_cur_y[i] = trees_y[trees_cur[i]]
    
    for i in range(0,4):
        trees_x[i]-=gameSpeed
    
    if ground_x1 == -1000:
        ground_x1 = 1000
    elif ground_x2 == -1000:
        ground_x2 = 1000
    ground_x1 -= gameSpeed
    ground_x2 -= gameSpeed
    
    if gamestop == False:
        game_x = frame(game_x,jump)
    else:
        gameDisplay.blit(trex1,(trex_x,trex_y))
        
    for i in range(4):
        if 50<trees_x[i]<150:
            if trees_cur_y[i]-trex_height+100<trex_y:
                gameSpeed = 0
                gamestop = True
    pygame.display.update()
    clock.tick(60)
    
pygame.quit()
```
