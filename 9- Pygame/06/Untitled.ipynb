```python
import pygame
import random
```

```python
def random_gap(edge,mini_gap,maxi_gap):
    return random.randint(edge+mini_gap,edge+maxi_gap)

def frame(game_x,jump):
    if 0<=game_x%10<5:
        if jump:
            gameDisplay.blit(trex_jump,(trex_x,trex_y))
        else:    
            gameDisplay.blit(trex1,(trex_x,trex_y))
    elif 5<=game_x%10<10:
        if jump:
            gameDisplay.blit(trex_jump,(trex_x,trex_y))
        else:
            gameDisplay.blit(trex2,(trex_x,trex_y))
    if game_x == 1000:
        game_x = 1
    else:
        game_x += 1
    return game_x
def show_message(textMessage, messageColor, messageFont, messagePosition_x,message_position_y):
    textMessage = str(textMessage)
    font_and_size = pygame.font.Font("freesansbold.ttf",messageFont)
    text_surface = font_and_size.render(textMessage,True,messageColor)
    text_rect = text_surface.get_rect()
    text_rect.center = ((messagePosition_x),(message_position_y))
    gameDisplay.blit(text_surface,text_rect)

def game_start():
    trees_cur = [random.randint(0,len(tree_image)-1),
                 random.randint(0,len(tree_image)-1),
                 random.randint(0,len(tree_image)-1),
                 random.randint(0,len(tree_image)-1)]
    trees_cur_y = [trees_y[trees_cur[0]],
                   trees_y[trees_cur[1]],
                   trees_y[trees_cur[2]],
                   trees_y[trees_cur[3]]
                  ]

    trees_x = [random.randint(trees_location[0]+min_gap,trees_location[0]+max_gap),
               random.randint(trees_location[1]+min_gap,trees_location[1]+max_gap),
               random.randint(trees_location[2]+min_gap,trees_location[2]+max_gap),
               random.randint(trees_location[3]+min_gap,trees_location[3]+max_gap)
              ]
    ground_x1 = 0
    ground_x2 = 1000
    game_x = 1
    score = 0
    score_rate = 0
    gamestop = False
    gameSpeed = 5
    
    return trees_cur, trees_cur_y, trees_x, ground_x1,ground_x2,game_x,score,gamestop,gameSpeed,score_rate
```

```python
pygame.init()

main_widnow_width = 1000
main_widnow_height = 500

min_gap = 0
max_gap = 50

gameDisplay = pygame.display.set_mode((main_widnow_width,main_widnow_height))
pygame.display.set_caption("Trex")

#Color
white = (255,255,255)
black = (0,0,0)
red = (255,0,0)

#Trees
tree_image = [pygame.image.load("tree_mix2.png"),
              pygame.image.load("tree1s1.png"),
              pygame.image.load("tree1s2.png"),
              pygame.image.load("tree1s3.png"),
              pygame.image.load("tree2s1.png"),
              pygame.image.load("tree2s2.png"),
              pygame.image.load("tree2s3.png"),
              pygame.image.load("tree3s1.png"),
              pygame.image.load("tree3s2.png"),
              pygame.image.load("tree3s3.png"),
              pygame.image.load("tree4s1.png"),
              pygame.image.load("tree4s2.png"),
              pygame.image.load("tree4s3.png"),]
 
trees_y = [295,300,320,345,300,320,345,300,320,345,300,320,345]

trees_width = [145,86,73,60,86,73,60,86,73,60,86,73,60]
trees_height = [130,130,110,90,130,110,90,130,110,90,130,110,90]

trees_location = [1000,1400,1800,2200]

trees_cur = [random.randint(0,len(tree_image)-1),
             random.randint(0,len(tree_image)-1),
             random.randint(0,len(tree_image)-1),
             random.randint(0,len(tree_image)-1)]
trees_cur_y = [trees_y[trees_cur[0]],
               trees_y[trees_cur[1]],
               trees_y[trees_cur[2]],
               trees_y[trees_cur[3]]
              ]

trees_x = [random.randint(trees_location[0]+min_gap,trees_location[0]+max_gap),
           random.randint(trees_location[1]+min_gap,trees_location[1]+max_gap),
           random.randint(trees_location[2]+min_gap,trees_location[2]+max_gap),
           random.randint(trees_location[3]+min_gap,trees_location[3]+max_gap)
          ]



#Trex
trex1 = pygame.image.load("trex_run1.png")
trex2 = pygame.image.load("trex_run2.png")
trex_jump = pygame.image.load("trex_jump.png")
trex_x = 50
trex_y = 300
trex_width = 233
trex_height = 216

#Ground
ground1 = pygame.image.load("desert.png")
ground_x1 = 0
ground_y1 = 350
ground2 = ground1
ground_x2 = 1000
ground_y2 = ground_y1

clock = pygame.time.Clock()

game_x = 1


jump = False
jump_i = -30

score = 0
score_rate = 60

gamespeed_rate = 0
game_speed_done = True

gamestop = False
gameSpeed = 5
alive = True
while alive:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            alive = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE and gamestop != True:
                if jump == False:
                    jump = True
        if event.type ==pygame.KEYDOWN:
            if event.key == pygame.K_RETURN:
                if gamestop:
                    trees_cur, trees_cur_y, trees_x, ground_x1,ground_x2,game_x,score,gamestop,gameSpeed,score_rate = game_start()
    if jump:
        if jump_i<=30:
            trex_y = (main_widnow_height-trex_height)- (180 - ((jump_i**2) / 5))
            jump_i+=1 
        else:
            jump = False
            jump_i = -30
    
    gameDisplay.fill(white)
    gameDisplay.blit(ground1,(ground_x1,ground_y1))
    gameDisplay.blit(ground2,(ground_x2,ground_y2))
    
    for i in range(0,4):
        gameDisplay.blit(tree_image[trees_cur[i]],(trees_x[i],trees_cur_y[i]))
        if trees_x[i]<-150:
            last_tree = max(trees_x)
            trees_x[i] = random.randint(1400+min_gap,1400+max_gap)
            trees_cur[i] = random.randint(0,len(tree_image)-1)
            trees_cur_y[i] = trees_y[trees_cur[i]]
            score+=10
    
    for i in range(0,4):
        trees_x[i]-=gameSpeed
    
    if ground_x1 <= -1000:
        ground_x1 = abs(ground_x1)
    elif ground_x2 <= -1000:
        ground_x2 =  abs(ground_x2)
    ground_x1 -= gameSpeed
    ground_x2 -= gameSpeed
    
    if gamestop == False:
        game_x = frame(game_x,jump)
        if score_rate == 60:
            score+=1
            score_rate = 0
    else:
        gameDisplay.blit(trex1,(trex_x,trex_y))
        
    for i in range(4):
        if 50<trees_x[i]<150:
            if trees_cur_y[i]-trex_height+100<trex_y:
                gameSpeed = 0
                gamestop = True
                show_message("Game Over", red, 80, main_widnow_width/2,main_widnow_height/2)
            
    show_message(score, black, 22, 950,11)
    
    score_rate += 0.5
    if game_speed_done:
        gamespeed_rate += 0.5

    if gamespeed_rate == 100 and game_speed_done == True :
        gameSpeed+=1
        gamespeed_rate = 0
        game_speed_done = False
    pygame.display.update()
    clock.tick(60)
    
pygame.quit()
```
