```python
import pygame
```

```python
pygame.init()

main_widnow_width = 1000
main_widnow_height = 500

gameDisplay = pygame.display.set_mode((main_widnow_width,main_widnow_height))
pygame.display.set_caption("Trex")

#Color
white = (255,255,255)

#Tree
tree = pygame.image.load("tree.png")
tree_x = 900
tree_y = 350
tree_width = 51
tree_height = 72

#Trex
trex1 = pygame.image.load("trex_run1.png")
trex2 = pygame.image.load("trex_run2.png")
trex_jump = pygame.image.load("trex_jump.png")
trex_x = 50
trex_y = 300
trex_width = 233
trex_height = 216

#Ground
ground1 = pygame.image.load("desert.png")
ground_x1 = 0
ground_y1 = 350
ground2 = ground1
ground_x2 = 1000
ground_y2 = ground_y1

clock = pygame.time.Clock()

game_x = 1

def frame(game_x,jump):
    if 0<=game_x%10<5:
        if jump:
            gameDisplay.blit(trex_jump,(trex_x,trex_y))
        else:    
            gameDisplay.blit(trex1,(trex_x,trex_y))
    elif 5<=game_x%10<10:
        if jump:
            gameDisplay.blit(trex_jump,(trex_x,trex_y))
        else:
            gameDisplay.blit(trex2,(trex_x,trex_y))
    if game_x == 1000:
        game_x = 1
    else:
        game_x += 1
    return game_x

jump = False
jump_i = -12

gameSpeed = 5

alive = True
while alive:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            alive = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                if jump == False:
                    jump = True
    if jump:
        if jump_i<=12:
            trex_y = (main_widnow_height-trex_height)- (144 - (jump_i**2))
            jump_i+=1
        else:
            jump = False
            jump_i = -12
    
    gameDisplay.fill(white)
    gameDisplay.blit(ground1,(ground_x1,ground_y1))
    gameDisplay.blit(ground2,(ground_x2,ground_y2))
    
    gameDisplay.blit(tree,(tree_x,tree_y))
    tree_x -= gameSpeed
    
    if ground_x1 == -1000:
        ground_x1 = 1000
    elif ground_x2 == -1000:
        ground_x2 = 1000
    ground_x1 -= gameSpeed
    ground_x2 -= gameSpeed
    game_x = frame(game_x,jump)
    pygame.display.update()
    clock.tick(60)
    
pygame.quit()
```
