```python
import cv2
```

```python
cap = cv2.VideoCapture(0)  # 0 or -1 (incase of more cameras use 1 2 3 etc.) you can also provide the address of a file

while True:
    ret, frame = cap.read()
    
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
