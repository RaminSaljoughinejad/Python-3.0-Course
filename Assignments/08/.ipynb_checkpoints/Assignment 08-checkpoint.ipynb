```python
while True:
    print("Please Enter 'Add' or 'add' to add two numbers")
    print("Please Enter 'Subtract' or 'subtract' to subtract two numbers")
    print("Please Enter 'Multiply' or 'moltiply' to multiply two numbers")
    print("Please Enter 'Divide' or 'divide' to divide two numbers")
    print("Please Enter 'Exit' or 'exit' to end the program.")
    print()
    try:
        userInput=str(input("Your Command: "))
        if userInput=="Add" or userInput=="add":
            print()
            print("Adding Two Numbers")
            try:
                x =float(input("Enter Your First Number: "))
                y =float(input("Enter Your Second Number: "))
                print()
                print("Answer: " + str(x+y))
                print("----------------------------------------------------------------------------")
            except ValueError:
                print()
                print("Error: Invalid Input - Only Numbers Are Allowed.")
                print("----------------------------------------------------------------------------")
                continue
        elif userInput=="Subtract" or userInput=="subtract":
            print()
            print("subtracting Two Numbers")
            try:
                x =float(input("Enter Your First Number: "))
                y =float(input("Enter Your Second Number: "))
                print()
                print("Answer: " + str(x-y))
                print("----------------------------------------------------------------------------")
            except ValueError:
                print()
                print("Error: Invalid Input - Only Numbers Are Allowed.")
                print("----------------------------------------------------------------------------")
                continue
        elif userInput=="Multiply" or userInput=="multiply":
            print()
            print("Multiplying Two Numbers")
            try:
                x =float(input("Enter Your First Number: "))
                y =float(input("Enter Your Second Number: "))
                print()
                print("Answer: " + str(x*y))
                print("----------------------------------------------------------------------------")
            except ValueError:
                print()
                print("Error: Invalid Input - Only Numbers Are Allowed.")
                print("----------------------------------------------------------------------------")
                continue
        elif userInput=="Divide" or userInput=="divide":
            print()
            print("Dividing Two Numbers")
            try:
                x =float(input("Enter Your First Number: "))
                y =float(input("Enter Your Second Number: "))
                print()
                try:
                    print("Answer: " + str(x/y))
                    print("----------------------------------------------------------------------------")
                except ZeroDivisionError:
                    print()
                    print("Error: Dividing By Zero is Not Allowed.")
                    print("----------------------------------------------------------------------------")
                    continue
            except ValueError:
                print()
                print("Error: Invalid Input - Only Numbers Are Allowed.")
                print("----------------------------------------------------------------------------")
                continue
        elif userInput=="Exit" or userInput=="exit":
            break
        else:
            print()
            print("Error: Invalid Input")
            print("----------------------------------------------------------------------------")
            continue
    except ValueError:
        print()
        print("Error: Invalid Input")
        print("----------------------------------------------------------------------------")
        continue
```
