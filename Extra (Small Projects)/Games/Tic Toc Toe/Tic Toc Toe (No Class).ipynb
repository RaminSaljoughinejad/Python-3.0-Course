```python
import pygame
```

```python
pygame.init()

game_display = pygame.display.set_mode((400,400))
pygame.display.set_caption("Tic Toc Toe")
game_bord_sq1 = pygame.draw.rect(game_display, (255,255,255),(6.25,6.25,125,125))
sq1_x, sq1_y = 6.25, 6.25
game_bord_sq2 = pygame.draw.rect(game_display, (255,255,255),(137.5,6.25,125,125))
sq2_x, sq2_y = 137.5, 6.25
game_bord_sq3 = pygame.draw.rect(game_display, (255,255,255),(268.75,6.25,125,125))
sq3_x, sq3_y = 268.75, 6.25

game_bord_sq4 = pygame.draw.rect(game_display, (255,255,255),(6.25,137.5,125,125))
sq4_x, sq4_y = 6.25, 137.5
game_bord_sq5 = pygame.draw.rect(game_display, (255,255,255),(137.5,137.5,125,125))
sq5_x, sq5_y = 137.5, 137.5
game_bord_sq6 = pygame.draw.rect(game_display, (255,255,255),(268.75,137.5,125,125))
sq6_x, sq6_y = 268.75, 137.5

game_bord_sq7 = pygame.draw.rect(game_display, (255,255,255),(6.25,268.75,125,125))
sq7_x, sq7_y = 6.25, 268.75
game_bord_sq8 = pygame.draw.rect(game_display, (255,255,255),(137.5,268.75,125,125))
sq8_x, sq8_y = 137.5, 268.75
game_bord_sq9 = pygame.draw.rect(game_display, (255,255,255),(268.75,268.75,125,125))
sq9_x, sq9_y = 268.75, 268.75

clock = pygame.time.Clock()

def clicker(turn,square_x, square_y, logic_board, point):
    if logic_board[point//3][point%3]==0:
        if turn:
            pygame.draw.line(game_display, (255,0,0), (square_x + 8, square_y + 8), (square_x + 125 - 8, square_y + 125 - 8),7)
            pygame.draw.line(game_display, (255,0,0), (square_x + 8, square_y + 125 - 8), (square_x + 125 - 8, square_y + 8),7)
            logic_board[point//3][point%3] = 1
        else:
            pygame.draw.circle(game_display, (0,0,255), (int(square_x + 62), int(square_y + 62)), 55, 7)
            logic_board[point//3][point%3] = 2
        return (not turn), logic_board
    else:
        return turn, logic_board

    
def winer_checker(game_logic_board):
    if game_logic_board[0][0] == game_logic_board[0][1] == game_logic_board[0][2] == 1:
        print("Player 1 won.")
        return True
    elif game_logic_board[0][0] == game_logic_board[0][1] == game_logic_board[0][2] == 2:
        print("Player 2 won.")
        return True
    elif game_logic_board[1][0] == game_logic_board[1][1] == game_logic_board[1][2] == 1:
        print("Player 1 won.")
        return True
    elif game_logic_board[1][0] == game_logic_board[1][1] == game_logic_board[1][2] == 2:
        print("Player 2 won.")
        return True
    elif game_logic_board[2][0] == game_logic_board[2][1] == game_logic_board[2][2] == 1:
        print("Player 1 won.")
        return True
    elif game_logic_board[2][0] == game_logic_board[2][1] == game_logic_board[2][2] == 2:
        print("Player 2 won.")
        return True
    elif game_logic_board[0][0] == game_logic_board[1][0] == game_logic_board[2][0] == 1:
        print("Player 1 won.")
        return True
    elif game_logic_board[0][0] == game_logic_board[1][0] == game_logic_board[2][0] == 2:
        print("Player 2 won.")
        return True
    elif game_logic_board[0][1] == game_logic_board[1][1] == game_logic_board[2][1] == 1:
        print("Player 1 won.")
        return True
    elif game_logic_board[0][1] == game_logic_board[1][1] == game_logic_board[2][1] == 2:
        print("Player 2 won.")
        return True
    elif game_logic_board[0][2] == game_logic_board[1][2] == game_logic_board[2][2] == 1:
        print("Player 1 won.")
        return True
    elif game_logic_board[0][2] == game_logic_board[1][2] == game_logic_board[2][2] == 2:
        print("Player 2 won.")
        return True
    elif game_logic_board[0][0] == game_logic_board[1][1] == game_logic_board[2][2] == 1:
        print("Player 1 won.")
        return True
    elif game_logic_board[0][0] == game_logic_board[1][1] == game_logic_board[2][2] == 2:
        print("Player 2 won.")
        return True
    elif game_logic_board[2][0] == game_logic_board[1][1] == game_logic_board[0][2] == 1:
        print("Player 1 won.")
        return True
    elif game_logic_board[2][0] == game_logic_board[1][1] == game_logic_board[0][2] == 2:
        print("Player 2 won.")
        return True
    return False
game_logic_board = [[0,0,0],
                    [0,0,0],
                    [0,0,0]]
turn = True
game_status = True
while game_status:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            game_status = False
        if event.type == pygame.MOUSEBUTTONUP:
            click_pos = pygame.mouse.get_pos()
            if game_bord_sq1.collidepoint(click_pos):
                turn, game_logic_board = clicker(turn,sq1_x, sq1_y, game_logic_board, 0)
            elif game_bord_sq2.collidepoint(click_pos):
                turn, game_logic_board = clicker(turn,sq2_x, sq2_y, game_logic_board, 1)
            elif game_bord_sq3.collidepoint(click_pos):
                turn, game_logic_board = clicker(turn,sq3_x, sq3_y, game_logic_board, 2)
            elif game_bord_sq4.collidepoint(click_pos):
                turn, game_logic_board = clicker(turn,sq4_x, sq4_y, game_logic_board, 3)
            elif game_bord_sq5.collidepoint(click_pos):
                turn, game_logic_board = clicker(turn,sq5_x, sq5_y, game_logic_board, 4)
            elif game_bord_sq6.collidepoint(click_pos):
                turn, game_logic_board = clicker(turn,sq6_x, sq6_y, game_logic_board, 5)
            elif game_bord_sq7.collidepoint(click_pos):
                turn, game_logic_board = clicker(turn,sq7_x, sq7_y, game_logic_board, 6)
            elif game_bord_sq8.collidepoint(click_pos):
                turn, game_logic_board = clicker(turn,sq8_x, sq8_y, game_logic_board, 7)
            elif game_bord_sq9.collidepoint(click_pos):
                turn, game_logic_board = clicker(turn,sq9_x, sq9_y, game_logic_board, 8)
    pygame.display.update()
    if winer_checker(game_logic_board):
        break
pygame.quit()
```
