```python
import pygame
import numpy as np
import random
```

```python
class Snake:
    
    WINDOW_WIDTH = 400
    WINDOW_HEIGHT = 400
    snake_img = pygame.image.load("slice_of_snake.png")
    food_img = pygame.image.load("food.png")
    
    def __init__(self):
        self.DEF_LOCATION = np.full((20,20),"") 
        self.DEF_LOCATION[10][2]="r"
        self.DEF_LOCATION[10][3]="r"
        self.DEF_LOCATION[10][4]="r"
        self.head = [10,4]
        self.tail = [10,2]
        self.food = [random.randint(0,9),random.randint(0,19)]
        self.DEF_LOCATION[self.food[0],self.food[1]]="f"
        self.location = self.DEF_LOCATION
        self.alive = True
    
    
    def update(self):
        if self.location[self.head[0]][self.head[1]] == "r":
            if self.location[self.head[0]][(self.head[1]+1)%20] == "":
                self.location[self.head[0]][(self.head[1]+1)%20] = "r"
                self.head[1]=(self.head[1]+1)%20
            elif self.location[self.head[0]][(self.head[1]+1)%20] == "f":
                food = self.new_food()
                self.location[self.head[0]][(self.head[1]+1)%20] = "r"
                self.location[self.head[0]][(self.head[1]+2)%20] = "r"
                self.head[1]=(self.head[1]+2)%20
                #self.head[1]=(self.head[1]+1)%20
            else:
                self.alive = False
        elif self.location[self.head[0]][self.head[1]] == "u":
            if self.location[(self.head[0]-1)%20][self.head[1]] == "":
                self.location[(self.head[0]-1)%20][self.head[1]] = "u"
                self.head[0]=(self.head[0]-1)%20
            elif self.location[(self.head[0]-1)%20][self.head[1]] == "f":
                food = self.new_food()
                self.location[(self.head[0]-1)%20][self.head[1]] = "u"
                self.location[(self.head[0]-2)%20][self.head[1]] = "u"
                self.head[0]=(self.head[0]-2)%20
                #self.head[0]=(self.head[0]-1)%20
            else:
                self.alive = False
        elif self.location[self.head[0]][self.head[1]] == "d":
            if self.location[(self.head[0]+1)%20][self.head[1]] == "":
                self.location[(self.head[0]+1)%20][self.head[1]] = "d"
                self.head[0]=(self.head[0]+1)%20
            elif self.location[(self.head[0]+1)%20][self.head[1]] == "f":
                food = self.new_food()
                self.location[(self.head[0]+1)%20][self.head[1]] = "d"
                self.location[(self.head[0]+2)%20][self.head[1]] = "d"
                self.head[0]=(self.head[0]+2)%20
                #self.head[0]=(self.head[0]+1)%20
            else:
                self.alive = False
        elif self.location[self.head[0]][self.head[1]] == "l":
            if self.location[self.head[0]][(self.head[1]-1)%20] == "":
                self.location[self.head[0]][(self.head[1]-1)%20] = "l"
                self.head[1]=(self.head[1]-1)%20
            elif self.location[self.head[0]][(self.head[1]-1)%20] == "f":
                food = self.new_food()
                self.location[self.head[0]][(self.head[1]-1)%20] = "l"
                self.location[self.head[0]][(self.head[1]-2)%20] = "l"
                self.head[1]=(self.head[1]-2)%20
                #self.head[1]=(self.head[1]-1)%20
            else:
                self.alive = False
        
        
        if self.location[self.tail[0]][self.tail[1]]=="r":
            self.location[self.tail[0]][self.tail[1]]=""
            self.tail[1]=(self.tail[1]+1)%20
        elif self.location[self.tail[0]][self.tail[1]]=="u":
            self.location[self.tail[0]][self.tail[1]]=""
            self.tail[0]=(self.tail[0]-1)%20
        elif self.location[self.tail[0]][self.tail[1]]=="l":
            self.location[self.tail[0]][self.tail[1]]=""
            self.tail[1]=(self.tail[1]-1)%20
        elif self.location[self.tail[0]][self.tail[1]]=="d":
            self.location[self.tail[0]][self.tail[1]]=""
            self.tail[0]=(self.tail[0]+1)%20
    
    
    def new_food(self):
        x = random.randint(0,19)
        y = random.randint(0,19)
        while self.location[x][y] != "":
            x = random.randint(0,19)
            y = random.randint(0,19)
        self.food = [x,y]
        self.location[x][y]="f"
            
        
    def move(self, direction):
        self.location[self.head[0]][self.head[1]]=direction

        
    def show(self):
        for i in range(20):
            for j in range(20):
                if self.location[i][j] == "f":
                    gameDisplay.blit(self.food_img,(j*20,i*20))
                elif self.location[i][j] != "":
                    gameDisplay.blit(self.snake_img,(j*20,i*20))
```

```python
while True:
    pygame.init()
    newGame = Snake()
    gameDisplay = pygame.display.set_mode((newGame.WINDOW_WIDTH, newGame.WINDOW_HEIGHT))
    pygame.display.set_caption("Snake")

    clock = pygame.time.Clock()

    while True:
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                newGame.alive=False
            if event.type==pygame.KEYDOWN:
                if event.key==pygame.K_UP:
                    newGame.move("u")
                elif event.key==pygame.K_DOWN:
                    newGame.move("d")
                elif event.key==pygame.K_LEFT:
                    newGame.move("l")
                elif event.key==pygame.K_RIGHT:
                    newGame.move("r")

        gameDisplay.fill((255,255,255))
        newGame.update()
        newGame.show()


        pygame.display.update()
        clock.tick(5)
        if not newGame.alive:
            print("Game Over")
            del newGame
            break
    x = input("Do you want to play again? ")
    if x=="y":
        continue
    else:
        pygame.quit()
        break
```
